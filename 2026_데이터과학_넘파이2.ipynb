```python
# 1. 라이브러리 임포트 및 랜덤 데이터 생성
import numpy as np # 넘파이(수치 계산 라이브러리) 임포트
import matplotlib.pyplot as plt # 맷플롯립(데이터 시각화 라이브러리) 임포트

# 1. 랜덤 데이터 생성
# 표준 정규 분포(평균 0, 표준편차 1)를 따르는 1000개의 랜덤 숫자 생성
ran_np = np.random.randn(1000)
```

```python
# 2. 랜덤 데이터 확인
# 생성된 랜덤 넘파이 배열 `ran_np`를 출력하여 내용을 확인합니다.
ran_np
```

```python
# 3. 랜덤 데이터 평균 계산
# `ran_np` 배열의 평균값을 계산하여 출력합니다.
ran_np.mean()
```

```python
# 4. 랜덤 데이터 표준편차 계산
# `ran_np` 배열의 표준편차를 계산하여 출력합니다.
ran_np.std()
```

```python
# 5. 랜덤 데이터 히스토그램
# `ran_np` 데이터의 분포를 히스토그램으로 시각화합니다.
# `bins=50`은 데이터를 50개의 구간으로 나누어 빈도수를 표시하라는 의미입니다.
plt.hist(ran_np,bins=50)
# sns.distplot(ran_np,fit=stats.norm)
```

```python
# 6. 랜덤 데이터 박스 플롯
# `ran_np` 데이터의 분포를 박스 플롯으로 시각화합니다.
# 박스 플롯은 데이터의 중앙값, 사분위수, 이상치를 보여줍니다.
plt.boxplot(ran_np)
```

```python
# 7. 정규 분포 데이터 생성 및 시각화
# 새로운 정규 분포를 위한 평균(mu)과 표준편차(sigma)를 정의합니다.
mu = 10  # 평균
sigma = 3  # 표준편차

# 기존 `ran_np` (표준 정규 분포)를 사용하여 새로운 정규 분포 `gauss`를 생성합니다.
# (X * sigma + mu 공식을 사용)
gauss = mu + sigma * ran_np  # gauss=mu+sigma*np.random.randn(1000)과 동일한 효과

# `gauss` 데이터의 히스토그램을 50개의 구간으로 나누어 시각화합니다.
plt.hist(gauss,bins=50)

# 원본 `ran_np` 데이터의 히스토그램을 같은 그래프에 시각화하여 비교합니다.
plt.hist(ran_np,bins=50)

# 그래프를 화면에 표시합니다.
plt.show()
```
